```python
import os
import time
from IPython.display import clear_output

Tunnel = "Ngrok"  #@param ["Ngrok", "Cloudflare", "LocalTunnel"]
ngrok_tunnel_authtoken = "31mg7nvuXw30OoLzwd7VrIJGVJy_4hSxtDNZHfEpjz94hArA3"  #@param {type:"string"}

file_path = "facefusion.py"

# 1️⃣ Remove 'share=True' from launch (needed for Colab/remote)
os.system(f"sed -i 's/launch(share=True,/launch(/g' {file_path}")

# 3️⃣ Set up tunnels
if Tunnel == "Ngrok":
    os.system("pip install -q pyngrok")
    from pyngrok import ngrok
    ngrok.set_auth_token(ngrok_tunnel_authtoken)
    http_tunnel = ngrok.connect(7860, bind_tls=True)
    clear_output()
    print("FaceFusion server is running!")
    print("Ngrok Tunnel Public URL:", http_tunnel.public_url)

elif Tunnel == "Cloudflare":
    os.system("curl -LO https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb")
    os.system("dpkg -i cloudflared-linux-amd64.deb")
    os.system("nohup cloudflared tunnel --url localhost:7860 > cf_out.log 2>&1 &")
    time.sleep(5)
    # Extract Cloudflare URL from log
    cloudflare_url = ""
    with open("cf_out.log", "r") as f:
        for line in f:
            if "trycloudflare.com" in line:
                cloudflare_url = line.strip()
                break
    clear_output()
    if cloudflare_url:
        print(f"FaceFusion server is running!\nCloudflare Tunnel Public URL: {cloudflare_url}")
    else:
        print("Cloudflare tunnel did not start yet. Check cf_out.log")

elif Tunnel == "LocalTunnel":
    os.system("npm install -g localtunnel")
    os.system("nohup lt --port 7860 > lt_out.log 2>&1 &")
    time.sleep(5)
    tunnel_url = ""
    with open("lt_out.log", "r") as f:
        for line in f:
            if "your url is:" in line.lower():
                tunnel_url = line.strip().split()[-1]
                break
    clear_output()
    if tunnel_url:
        print(f"FaceFusion server is running!\nLocalTunnel Public URL: {tunnel_url}")
    else:
        print("LocalTunnel did not start yet. Check lt_out.log")

!python facefusion.py run
```
